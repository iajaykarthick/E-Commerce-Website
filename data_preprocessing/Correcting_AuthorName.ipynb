```python
import pandas as pd 
import numpy as np 
from ast import literal_eval
```

```python
books = pd.read_csv('BX-Books.csv', escapechar='\\', sep =';', encoding='latin-1')
books.shape
```

```python
## remove Publishing in author part 
```

```python
books[books['Book-Author'].str.contains('Education', na= False)]
```

```python
books[books['Book-Author'].str.contains('Publishing', na= False)]
```

```python
## Getting the publishing and education present in the author column 
education_index = books[books['Book-Author'].str.contains('Education', na= False)].index 
books =  books.drop(education_index)
publishing_index = books[books['Book-Author'].str.contains('Publishing', na= False)].index
books = books.drop(publishing_index)
```

```python
# dropping where the authors name is just 1 to 3 words
author_with_3words = books[books['Book-Author'].str.contains('^\w{1,3}\s&amp;', regex=True, na= False)].index
books = books.drop(author_with_3words)
```

```python
two_authors = books[books['Book-Author'].str.contains('&amp;', na= False)].copy()
```

```python
two_authors
```

```python
authors = []
for i in two_authors.index:
    a = list(map(lambda a:a.strip(), two_authors['Book-Author'][i].split('&amp;')))
    authors.append(a)
```

```python
two_authors['authorslist'] = authors
two_authors.head()
```

```python
#Creating Duplicates 
two_authors2 = pd.concat([two_authors]*2, ignore_index=True).copy()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# In original row keeping author first name and in duplicate keeping second authors name 
a = np.empty((int(len(two_authors2)/2)),dtype = "<U10")
count =0 
for i in range(len(two_authors2)):
    
    if two_authors2.iloc[i,0] not in a:
        a[count] = two_authors2.iloc[i,0]
        count +=1
        two_authors2.iloc[i,2] = two_authors2.iloc[i,-1][0]
        
    else:
        two_authors2.iloc[i,2] = two_authors2.iloc[i,-1][1]
```

```python
two_authors2[two_authors2['ISBN'] =='0061076031' ]
```

```python
two_authors.index
```

```python
books2 = books.drop(two_authors.index).copy()
books2.head()
```

```python
books2 = pd.concat([books2, two_authors2],ignore_index=True)
books2.drop(columns=['authorslist'],inplace=True)
```

```python
# getting author names who have issue with their names 
name_commaissue = books2[books2['Book-Author'].str.contains(',',na=False)].index
```

```python
books2.drop(name_commaissue,inplace=True)
books2.reset_index(inplace=True)
books2.drop(columns=['index'],inplace=True)
```

```python
#Droping authors with just 2 letters in their name
author_with_2letters = books2[books2['Book-Author'].str.contains('^\w{1,2}$', regex=True, na= False)].index
books2.drop(author_with_2letters,inplace=True)
books2.reset_index(inplace=True)
books2.drop(columns=['index'],inplace=True)
```

```python
books2.to_csv('Books_CorrectAuthorName.csv')
```
