```python
isbn_list = books.ISBN.unique()
```

```python
def get_html(ISBN):
    URL = f"https://www.goodreads.com/search?q={ISBN}"
    r = requests.get(URL)  
    html_soup = BeautifulSoup(r.content, 'html.parser') 
    return html_soup
```

```python
count = 1
desc_dict = {isbn: '' for isbn in isbn_list}
for isbn in isbn_list:
    desc_div = None
    trial = 5
    while not desc_div and trial > 0:
        trial -= 1
        html = get_html(isbn)
        desc_div = html.find('div', {'id': 'description'})        
    if desc_div:
        if desc_div.find_all(attrs={"style" : "display:none"}):
            print(count)
            print(isbn)
            desc_dict[isbn] = desc_div.find_all(attrs={"style" : "display:none"})[0].text
            count += 1
```

```python
isbn_l = []
desc_l = []
for isbn in desc_dict:  
    isbn_l.append(isbn)
    desc_l.append(desc_dict[isbn])
```

```python
desc_df = pd.DataFrame()
desc_df['isbn'] = isbn_l
desc_df['desc'] = desc_l
```

```python
books_img_df = pd.read_csv('raw_data/Books_CorrectAuthorName.csv', usecols=['ISBN', 'Image-URL-S', 'Image-URL-M','Image-URL-L'])
books_img_df = books_img_df.drop_duplicates()
books_img_df
```

```python
books_df = pd.read_csv('books.csv')
books_df
```

```python
desc_df = desc_df.rename(columns={'isbn': 'ISBN'})
desc_df
```

```python
final_books_df = pd.merge(pd.merge(books_df, books_img_df, on="ISBN"), desc_df, on="ISBN")
final_books_df
```

```python
final_books_df = final_books_df.apply(lambda x: x.str.replace('"', '').str.replace('`', '').str.replace("'", "") if x.name == 'desc' else x)
```

```python
final_books_df.rename(columns={'Book-Title': 'title', 'Price': 'price', 'Year-Of-Publication': 'published_year', 'Publisher ID': 'publisher_id', 'Image-URL-S':'image_short', 'Image-URL-M':'image_medium', 'Image-URL-L':'image_large'}, inplace=True)
final_books_df
```

```python
final_books_df[['ISBN', 'title', 'desc', 'price', 'publisher_id', 'published_year', 'image_short', 'image_medium', 'image_large']].to_csv('final_books.csv', index=False)
```
