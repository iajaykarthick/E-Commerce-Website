```python
import pandas as pd
```

Downloaded rawdata from https://www.unitedstateszipcodes.org/zip-code-database/

```python
zip = pd.read_csv('raw_data/zip_code_database.csv')
zip.head()
```

```python
zip.country.value_counts()
```

ZIP Codes range from 00501, belonging to the Internal Revenue Service in Holtsville, NY, to 99950 in Ketchikan, AK.

```python
zip = zip[zip['country'] == 'US']
zip.shape
```

```python
zip.isna().sum()
```

```python
zip = zip[['zip', 'primary_city', 'state', 'country']]
```

```python
zip.isna().sum()
```

```python
zip['zipcode'] = zip['zip'].astype(str)
```

```python
zip[zip['zipcode'].str.contains('^\d{,4}$', regex=True)]
```

```python
zip['zipcode'] = zip['zipcode'].str.zfill(5)
zip['zipcode']
```

```python
zip = zip.rename(columns = {'primary_city': 'city'})
zip = zip[['zipcode', 'city', 'state', 'country']]
zip.head()
```

```python
zip.to_csv('preprocessed_data/zip_code.csv')
```
